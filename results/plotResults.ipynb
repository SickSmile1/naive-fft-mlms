```python
%matplotlib notebook
plot1 = np.loadtxt("output_mlms1", delimiter=",",skiprows=1, usecols=(1,3))
plot2 = np.loadtxt("output_mlms2", delimiter=",",skiprows=1, usecols=(1,3))
plot3 = np.loadtxt("output_fft_8", delimiter=",",skiprows=1, usecols=(1,3))
plot1 = plot1*.001
plot2 = plot2*.001
plot3 = plot3*.001
t = np.array([8,16,32,64,128,256,512,1024,2048,4096])
t = t**2
plt.plot(t,t*np.log(t)*1e-7, color="red")
plt.plot(t,plot1[:,1],t,plot2[:,1],t,plot3[:,1])
plt.yscale("log")
plt.xscale("log")
plt.ylabel("Time")
plt.xlabel("Grid")
plt.title("MLMS with t=3")
plt.legend(["Optimized MLMS","MLMS","FFT"])
plt.show()
#plot1[:,0]
```

```python
%matplotlib notebook
plot1_t = np.loadtxt("output_mlms1_t2", delimiter=",",skiprows=1, usecols=(1,3))
plot2_t = np.loadtxt("output_mlms2_t2", delimiter=",",skiprows=1, usecols=(1,3))
plot3_t = np.loadtxt("output_fft_8", delimiter=",",skiprows=1, usecols=(1,3))
plot1_t = plot1_t*.001
plot2_t = plot2_t*.001
plot3_t = plot3_t*.001
plt.plot(t,t*np.log(t)*1e-7, color="red")
plt.plot(t,t*1e-7, color='purple')
plt.plot(t,plot1_t[:,1],t,plot2_t[:,1],t,plot3_t[:,1])
#plt.yscale("log")
#plt.xscale("log")
plt.ylabel("Time")
plt.xlabel("Grid")
plt.title("MLMS with t=2")
plt.legend(["Optimized MLMS","MLMS","FFT"])
plt.show()
```
