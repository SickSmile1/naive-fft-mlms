```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
arr = np.fromfile("60_grid", sep='\t')
arr2 = arr.reshape(60,60)

surf = plt.imshow(arr2,cmap='viridis')
plt.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
x,y = np.loadtxt("timings_home.txt",
                 delimiter='\t',usecols=(0, 1), unpack=True)
fig, ax = plt.subplots()
print(x)
x2 = np.power(x,2)

#ax.set_yscale("log")
#ax.set_xscale('log')
plt.xlabel("number of nodes")
plt.ylabel("time in seconds")
plt.grid(color='r', alpha=0.2, linestyle='-', linewidth=.5)
ax.plot(x2,y)

#ax.set_aspect("equal")
```
